```python
import pandas as pd
import numpy as np
```

```python
subway_xy = pd.read_csv('./station_coordinate.csv')
subway_ko = pd.read_excel('./코레일주소.xlsx')
subway_se = pd.read_csv('./서울교통공사_역주소 및 전화번호_20220110.csv', encoding='cp949')
```

```python
subway_se = subway_se[['역명', '구주소']]
subway_ko = subway_ko[['역명', '지번주소']]
```

```python
subway_se.head()
```

```python
subway_ko.head()
```

```python
subway_se.isnull().sum()
```

```python
subway_ko[subway_ko['주소'].isna()]
```

```python
subway_se = subway_se.rename(columns={'구주소' : '주소'})
subway_ko = subway_ko.rename(columns={'지번주소':'주소'})
```

```python
subway_ = pd.concat([subway_se, subway_ko], axis=0)
```

```python
subway_.isnull().sum()
```

```python
subway_[subway_['주소'].isnull()]
```

```python
subway_.loc[subway_['역명']=='광명','주소'] = '경기도 광명시 일직동'
```

```python
subway_.dropna()
```

```python
subway = subway_[subway_['주소'].str.contains('서울') | subway_['주소'].str.contains('경기')]
```

```python
subway.head()
```

```python
subway_splited = subway['주소'].str.split(' ', expand=True)
subway_splited = subway_splited.drop([3,4,5], axis=1)
```

```python
subway = pd.concat([subway_splited, subway], axis=1)
subway = subway.drop(['주소'], axis=1)
```

```python
subway.columns = ['서울/경기','시/군/구','동','역명']
```

```python
subway['서울/경기'].unique()
```

```python
subway['서울/경기'] = subway['서울/경기'].str.replace('서울툭별시','서울특별시')
subway['서울/경기'] = subway['서울/경기'].str.replace('서울특셜시','서울특별시')
```

```python
subway['서울/경기'].unique()
```

```python
seoul_subway = subway[subway['서울/경기'].str.contains('서울')]
gg_subway = subway[subway['서울/경기'].str.contains('경기')]
```

```python
address = seoul_subway['시/군/구'] + ' ' + seoul_subway['동']

count_by_address = address.value_counts()

df = pd.DataFrame({'시군구동': count_by_address.index, '역 개수': count_by_address.values})
df
```

```python
address_2 = gg_subway['시/군/구'] + ' ' + gg_subway['동']

count_by_address_2 = address_2.value_counts()

df_2 = pd.DataFrame({'시군구동': count_by_address_2.index, '역 개수': count_by_address_2.values})
df_2
```

```python
seoul_ = pd.read_csv('./seoul_sb_처리중_4.csv')
gg_ = pd.read_csv('./gg_sb_처리중_4.csv')
```

```python
seoul_['시군구동'] = seoul_['시/군/구'] + ' ' + seoul_['동']
gg_['시군구동'] = gg_['시/군/구'] + ' ' + gg_['동']
seoul_ = pd.merge(seoul_, df, left_on='시군구동', right_on='시군구동').drop(['시군구동'], axis=1)
gg_ = pd.merge(gg_, df_2, left_on ='시군구동', right_on = '시군구동').drop(['시군구동'], axis=1)
seoul_.sample(n=5)
```
