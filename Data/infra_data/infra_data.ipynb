### 데이터 만져보기

1. 필요없는 column들 drop하기
2. 필요한 정보만 뽑아보기 starbucks에 대한 정보

```python
starbucks_csv.dtypes
```

```python
sb_df = starbucks_csv[['사업장명','상세영업상태명', '인허가일자', '폐업일자', '좌표정보(X)', '좌표정보(Y)']]
sb_df = sb_df[sb_df['사업장명'].str.contains('스타벅스')]
sb_df.reset_index(drop=True, inplace=True)
sb_df
```
